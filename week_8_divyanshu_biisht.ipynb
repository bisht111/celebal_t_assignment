# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator
import re
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""


    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
import logging
from datetime import datetime

# Configure logging
logging.basicConfig(
    filename="agent.log",      # Saves logs to agent.log
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    force=True
)

def log_agent(action, query, response):
    """
    Logs every agent request and response.
    """
    logging.info(
        f"Action: {action} | Query: {query} | Response: {response}"
    )
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python

def agent(query: str):
    query_lower = query.lower()

    try:
        if "calculate" in query_lower:
            expression = re.sub(r"[^0-9+\-*/(). ]", "", query)

            result = calculator(expression)

            if result == "Error in calculation":
                response = {
                    "type": "error",
                    "result": result
                }
                log_agent("Calculation Error", query, response)
                log_agent("Calculation", query, response)
                return response

            response = {
                "type": "calculation",
                "result": result,
            }

            log_agent("Calculation", query, response)
            return response

        elif "keywords" in query_lower:
            text = query.replace("Extract keywords from", "")
            text = text.replace("extract keywords from", "").strip()

            response = {
                "type": "keywords",
                "result": extract_keywords(text)
            }

            log_agent("Keyword Extraction", query, response)
            return response

        else:
            response = {
                "type": "general",
                "result": f"I received your query: '{query}'. This is a general response."
            }

            log_agent("Error", query, response)
            return response

    except Exception as e:
        response = {
            "type": "error",
            "result": str(e)
        }

        log_agent("Exception", query, response)
        return response

    return {
        "type": "unknown",
        "result": "Not implemented"
    }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```
