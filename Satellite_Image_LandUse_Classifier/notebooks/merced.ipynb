```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
```

```python
import torch

from src.models import ResNet18Classifier
from src.uc_merced import get_uc_merced_loader
```

```python
UC_DIR = "/home/divye/Desktop/cnn/data/UCMerced_LandUse/Images"

loader, uc_classes = get_uc_merced_loader(
    UC_DIR,
    batch_size=32
)
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

model = ResNet18Classifier(10)

model.load_state_dict(
    torch.load(
        "../checkpoints/resnet18_best.pt",
        map_location=device
    )
)

model.to(device)

model.eval()
```

```python
predictions = []
labels = []

with torch.no_grad():

    for images, target in loader:

        images = images.to(device)

        outputs = model(images)

        pred = outputs.argmax(dim=1).cpu().numpy()

        predictions.extend(pred)

        labels.extend(target.numpy())

print("Inference complete.")
```
