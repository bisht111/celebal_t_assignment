```python
import os
import random
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
```

```python
from pathlib import Path

DATA_DIR = Path("/home/divye/Desktop/cnn/data/2750")

print(DATA_DIR.exists())
```

```python
classes = sorted([folder.name for folder in DATA_DIR.iterdir() if folder.is_dir()])

print("Classes:")
print(classes)

print("\nNumber of classes:", len(classes))
```

```python
class_counts = {}

for cls in classes:
    class_path = DATA_DIR / cls
    images = list(class_path.glob("*.jpg"))
    class_counts[cls] = len(images)

df = pd.DataFrame({
    "Class": class_counts.keys(),
    "Images": class_counts.values()
})

df
```

```python
bad_images = []

for cls in classes:
    images = list((DATA_DIR / cls).glob("*.jpg"))

    for img_path in images:
        try:
            img = Image.open(img_path)
            img.verify()
        except Exception:
            bad_images.append(img_path)

print("Corrupted Images:", len(bad_images))
```

```python
print(f"Total Classes : {len(classes)}")
print(f"Total Images  : {sum(class_counts.values())}")
print(f"Average/Class : {sum(class_counts.values())/len(classes):.1f}")
```
